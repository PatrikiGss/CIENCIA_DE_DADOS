NumPy

NumPy (Numerical Python - Python Numérico) é uma biblioteca de computação científica que fornece suporte para arrays multidimensionais (ndarray), bem como funções matemáticas e de leitura e escrita de arquivos.

Para usá-lo em um programa Python, deve-se importar o módulo no início do programa.

```python
import numpy as np
```

ndarray

Um ndarray é um conteiner multidimensional homogêneo, ou seja, diferentemente das listas padrões do Python, apenas valores de um mesmo tipo podem ser armazenados dentro dele.

A criação de um ndarray pode ser feita com as seguintes funções:

array(): cria um array a partir de uma lista;

```python
#Criação de ndarray usando o comando array, a partir de listas.
lista = [1,2,3,4,5]
arr = np.array(lista)
print("Lista: ", lista)
print("ndarray: ", arr)
```

arange(): cria um array com valores dentro de um intervalo e de um passo especificado;

```python
#especifica apenas a quantidade de elementos (inicia em zero e vai até n - 1)
arr = np.arange(10)
print("ndarray: ", arr)

#especifica o valor inicial e o final
arr = np.arange(3,8)
print("ndarray: ", arr)

#especifica o valor inicial, o final e o passo
arr = np.arange(1,10,2)
print("ndarray: ", arr)
```

zeros(): cria um array de uma determinada dimensão preenchido de zeros;

```python
arr = np.zeros(5)
print("ndarray com uma dimensão: ")
print(arr)

arr = np.zeros((5,5))
print("ndarray com duas dimensões: ")
print(arr)

arr = np.zeros((3,5,5))
print("ndarray com três dimensões: ")
print(arr)
```

ones(): cria um array de uma determinada dimensão preenchido de uns;

```python
arr = np.ones(5)
print("ndarray com uma dimensão: ")
print(arr)

arr = np.ones((5,5))
print("ndarray com duas dimensões: ")
print(arr)
```

full(): cria um array de uma determinada dimensão preenchido de com um valor constante.

```python
arr = np.full((5),3)
print("ndarray com uma dimensão: ")
print(arr)

arr = np.full((5,5),18)
print("ndarray com duas dimensões: ")
print(arr)
```

eye(): cria uma matriz identidade;

```python
arr = np.eye(3)
print("Matriz identidade com três linhas e três colunas: ")
print(arr)
```

random(): cria um array de uma determinada dimensão preenchido com números aleatórios.

```python
arr = np.random.random(5)
print("ndarray com uma dimensão: ")
print(arr)

arr = np.random.random((3,3))
print("ndarray com duas dimensões: ")
print(arr)
```

Um ndarray possui um shape que é definido durante a sua criação e representa a quantidade de dimensões e elementos em cada dimensão do ndarray. Esta informação, assim como o seu tipo podem, ser obtidos usando os atributos shape e dtype.

```python
arr = np.random.random((3,3))
print("ndarray com duas dimensões: ")
print(arr)
print(arr.shape)
print(arr.dtype)
```

A função reshape possibilita que a forma de um array seja modificada, sem prejuízo em seus dados.

```python
a = np.random.random(6)
print("ndrray original:")
print(a)
b = a.reshape(2,3)
print("ndrray alterado para matriz 2x3:")
print(b)
```

ndarray podem ser indexados como como os vetores e matrizes das maioria das linguagens de programação.

```python
arr = np.random.random((3,3))
print(arr)
print("Elemento da linha 0 coluna 1:",arr[0,1])
print("Elemento da linha 2 coluna 2:",arr[2,2])
arr[1,1] = 0
print("ndarray com o valor da posição 1,1 alterado para 0.")
print(arr)
```

Os tipos numéricos que podem ser usados em um vetor NumPy (ndarray) são apresentados na tabela abaixo. Eles podem ser passador por parâmetros para as funções que fazem a criação dos vetores.

TiposNumPy.png

```python
arr = np.eye(3)
print("Matriz identidade com três linhas e três colunas e valores reais: ")
print(arr)

arr = np.eye(3, dtype=np.int64)
print("Matriz identidade com três linhas e três colunas e valores inteiros: ")
print(arr)

arr = np.eye(3, dtype=np.bool_)
print("Matriz identidade com três linhas e três colunas e valores booleanos: ")
print(arr)
```

Uma das grandes vantagens dos vetores NumPy é que eles permitem que as operações aritméticas sejam feitas diretamente, sem a necessidade de estruturas de repetição. É preciso cuidar para que a forma (shape) dos dois vetores sejam iguais.

```python
a = np.arange(10)
print("ndarray original: ")
print(a)
b = 2 * a
print("ndarray a multiplicado por 2: ")
print(b)

a = np.eye(3)
print("Matriz identidade com três linhas e três colunas e valores original: ")
print(a)
b = a + 3
print("Matriz identidade com três linhas e três colunas mais 3: ")
print(b)
```

Operações de comparação podem ser feitas também para os vetores inteiros.

```python
a = np.arange(1,10)
print("ndarray a: ")
print(a)

b = np.full(9,5)
print("ndarray b: ")
print(b)

c = a < b
print(c)
```

Também é possível usar operadores de comparação e lógicos para indexar vetores

```python
a = np.arange(0,10)
print("ndarray a: ")
print(a)
b = np.random.random(10)
print("ndarray b: ")
print(b)
c = b[a>5]
print("ndarray c: ")
print(c)
```

Slices

O fatiamento de dados em arrays unidimensionais do NumPy é semelhante a listas em Python. Entretanto, por questões de otimização, os slices são referências para o array original, portanto ao alterar um valor do array gerado por um slice, também se altera o valor original.

```python
a = np.arange(1,10)
print("ndarray a: ")
print(a)

b = a[3:6]
print("ndarray b: ")
print(b)

b[1] = 10

print("ndarray a: ")
print(a)
print("ndarray b: ")
print(b)
```

Retornando às listas em Python é possível perceber que elas se comportam de forma diferente dos arrays do NumPy

```python
a = list(range(1,10))
print("Lista a: ")
print(a)

b = a[3:6]
print("Lista b: ")
print(b)

b[1] = 10

print("Lista a: ")
print(a)
print("Lista b: ")
print(b)
```

Para usar vetores NumPy e evitar o problema anterior, deve-se usar a função copy.

```python
a = np.arange(1,10)
print("ndarray a: ")
print(a)

b = a[3:6].copy()
print("ndarray b: ")
print(b)

b[1] = 10

print("ndarray a: ")
print(a)
print("ndarray b: ")
print(b)
```

A diferença na forma como os arrays e listas são implementados, assim como as funções que os manipulam trazem diferenças muito significativas em termos de tempo de processamento.

```python
import time

ti = time.time()

lista = list(range(1000000))
for i in range(1000000):
  lista[i] = lista[i] * 2
print(lista[100000])
print(lista[999999])

tf = time.time()
tempo = tf - ti
print("Tempo para execução:",tempo)
```

```python
import time

ti = time.time()

array = np.arange(1000000)
array = array * 2
print(array[100000])
print(array[999999])

tf = time.time()
tempo = tf - ti
print("Tempo para execução:",tempo)
```

Funções vetorizadas

São funções que são executadas em todos os elementos de um array, sem a necessidade de percorrê-lo explicitamente com uma estrutura de repetição. Dentre outras, destacam-se:

abs: valor absoluto;
sqrt: raiz quadrada;
square: eleva ao quadrado;
exp: e elevado a x;
log, log10, log2, log1p: calcula o logaritmo nas diferentes bases (1p = 1 + p);
ceil, floor, rint, round: funções de arredondamento;
modf: devolve as partes inteira e fracionária em vetores separados;
isnan: retorna um valor de booleans indicando se o elemento de cada posição é um número ou não;
cos, cosh, sin, sinh, tan, tanh, arccos, arcosh, arcsin, arcsinh, arctan, arctanh: funções trigonométricas.
power: eleva os elementos do primeiro vetor às potências especificadas no segundo vetor.

```python
a = np.arange(10)
print("Vetor original: ")
print(a)

b = np.sqrt(a)
print("Raiz quadrada: ")
print(b)

c = np.round(b,2)
print("Raiz arredondada para duas casas: ")
print(c)

d = np.square(a)
print("Elevado ao quadrado: ")
print(d)
```
