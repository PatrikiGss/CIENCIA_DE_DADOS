```python
import pandas as pd
import numpy as np
```

Leia o arquivo CSV disponível na página da disciplina e apresente o DataFrame completo.

na_values é para que strings vazias virem not a NAN e sejam lidas pelas funções do pandas

```python
dados = pd.read_csv("Aula1_Atividade1.csv", delimiter=",", na_values=[''])
dados = dados.set_index('Nome')
```

Transforme a coluna nome em índice e reapresente o DataFrame.

```python
print("DataFrame com a coluna Nome como índice")
print(dados)
```

Conte quantos elementos nulos existem em cada coluna.

```python
print("\nNulos por coluna:")
print(dados.isnull().sum())

print("\nTotal de nulos no DataFrame:")
print(dados.isnull().sum().sum())
    
```

Elimine as linhas com elementos nulos do dataframe.

```python
dados_sem_nulos = dados.dropna()
print("\nDataFrame sem as linhas com valores nulos:")
print(dados_sem_nulos)
```

Mostre os dados da pessoa com nome Francisco.

```python
linha = dados_sem_nulos.loc['Francisco']
print("Todos os dados da linha do Francisco")
print(linha)
```

Calcule o Índice de Massa Corporal, usando a seguinte fórmula IMC = Peso / Altura**2. O valor do IMC deve ser salvo 

```python
dados_sem_nulos['Indice']= dados_sem_nulos['Peso']/ dados_sem_nulos['Altura']**2
print(dados_sem_nulos)
```

Mostre o nome e IMC dos homens obesos (IMC maior que 25).


```python
homens_obesos = dados_sem_nulos[(dados_sem_nulos['Genero'] == 'Masculino') & 
                                (dados_sem_nulos['Indice'] >= 25)]
print(homens_obesos)
```


Mostre o nome e IMC das mulheres obesas (IMC maior que 24).

```python
mulheres_obesas = dados_sem_nulos[(dados_sem_nulos['Genero'] == 'Feminino') & 
                                (dados_sem_nulos['Indice'] >= 24)]
print(mulheres_obesas)
```
