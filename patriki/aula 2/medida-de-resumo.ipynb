Cálculo das estatísticas básicas do atributo idade usando as funções do Pandas

```python
quantidade =  df['Idade'].count()
print('Quantidade de elementos: ', quantidade)

soma = df['Idade'].sum()
print('Soma dos elementos: ', soma)

media = df['Idade'].mean()
print('Media dos elementos: ', media)

mediana = df['Idade'].median()
print('Mediana dos elementos: ', mediana)

quartil1 = df['Idade'].quantile(q=0.25)
print('Quartil 1 dos elementos: ', quartil1)

quartil2 = df['Idade'].quantile(q=0.50)
print('Quartil 2 (mediana) dos elementos: ', quartil2)

quartil3 = df['Idade'].quantile(q=0.75)
print('Quartil 3 dos elementos: ', quartil3)

moda = df['Idade'].mode()
print('Moda dos elementos: ')
for x in moda:
  print(x)

minimo = df['Idade'].min()
print('Menor valor: ', minimo)

maximo = df['Idade'].max()
print('Maior valor: ', maximo)

var = df['Idade'].var()
print('Variância dos elementos: ', var)

dp = df['Idade'].std()
print('Desvio padrão dos elementos: ', dp)

cv = dp / media * 100
print('Coeficiente de variação dos elementos: ', cv)

assimetria = df['Idade'].skew()
print('Coeficiente de assimetria: ', assimetria)

curtose = df['Idade'].kurtosis()
print('Coeficiente de Curtose: ', curtose)
```

Parte 2 - Dados multidimensionais

Cálculo da covariância e da correlação entre as variáveis quantitativas do Dataframe.

Ao calcular o coeficiente de correlação deve-se indicar o tipo de cálculo a ser feito, por meio do parâmetro method, que pode receber os seguintes valores:

pearson (padrão); kendall; spearman.

```python
df.cov(numeric_only=True)
```

```python
df.corr(numeric_only=True)
```

```python
df.corr(method='kendall',numeric_only=True)
```

```python
df.corr(method='spearman',numeric_only=True)
```

Dados multidimensionais categóricos e numéricos

Cálculo do salário médio de cada função.

Todas as funções estatísticas usadas para o tratamento de dados numéricos unidimensionais podem ser usadas para calcular os valores de cada um dos valores de um atributo numérico, após agrupar o dataset com base no atributo categórico.

```python
dfAgrupadoFuncao = df.groupby(['Função'])

salarioAgrupadoFuncao = dfAgrupadoFuncao['Salário'].mean()

salarioAgrupadoFuncao
```

```python
salarioAgrupadoFuncao = df.groupby(['Função'])['Salário'].mean()

salarioAgrupadoFuncao
```

O Agrupamento pode ser realizado por mais de um atributo, como nos exemplos a seguir.

Inserção no dataframe de um novo atributo, denominado Gênero, com os valores armazenados na lista.

```python
df['Gênero'] = ['M','F','M','F','M','F','M','F','M','F','M','F','M','F','M','F','M','F','M','F']

df
```

```python
dfAgrupadoFuncaoGenero = df.groupby(['Função','Gênero'])
salarioAgrupadoFuncaoGenero = dfAgrupadoFuncaoGenero['Salário'].mean()
salarioAgrupadoFuncaoGenero
```

```python
dfAgrupadoGeneroFuncao = df.groupby(['Gênero','Função'])
salarioAgrupadoGeneroFuncao = dfAgrupadoGeneroFuncao['Salário'].mean()
salarioAgrupadoGeneroFuncao
```
