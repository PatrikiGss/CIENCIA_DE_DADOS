Elimine a coluna Unnamed: 0.

Verifique quantos valores ausentes (nulos) existem em cada atributo e elimine as linhas que contenham algum valor nulo.(dados = dados.dropna(axis=1)
 para dropar toda uma coluna com nulos)

```python
dados = df.dropna()

print("\nApós remoção, valores nulos por coluna:")
print(dados.isnull().sum())
```

Conte a quantidade de linhas duplicadas e elimine estas linhas.(df.duplicated(subset=['coluna1', 'coluna2']).sum() para se for verificar colunas em especifico)

```python
print(dados.duplicated().sum())

# Remove as linhas duplicadas (mantém apenas a primeira ocorrência)
dados.drop_duplicates(inplace=True)

# Confirma se ainda há duplicadas
print("Após remoção:", dados.duplicated().sum())
```

Apresente e depois elimine os outliers do campo salário. Considere que um outlier será dado pelo valor da média +- 2 vezes o seu desvio padrão.

```python
# calcular limites
li = df['Salary'].mean() - 2 * df['Salary'].std()
ls = df['Salary'].mean() + 2 * df['Salary'].std()

# exibir os outliers
outliers = df[(df['Salary'] < li) | (df['Salary'] > ls)]
print(outliers)

# eliminar os outliers
df_sem_outliers = df[(df['Salary'] >= li) & (df['Salary'] <= ls)]
```
